```python
X = df[['Product List Price', 'Product Category','Size of Store']]
```

```python
y = df[['Item_Outlet_Sales']]
```

Converting categorical variable into dummy/indicator variables and drop one in each category:

```python
X = pd.get_dummies(data=X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .20, random_state = 50)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(X,y)
```

```python
y_predict = reg.predict(X_test)
```

```python
y_predict
```

```python
#R2
score = reg.score(X, y)
print(f"Coefficient of determination R2 is {score.mean()*100:.2f} %")
```

```python
#The mean squared error (MSE)
from sklearn.metrics import mean_squared_error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_predict))
```
