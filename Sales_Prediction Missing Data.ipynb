```python
import pandas as pd
import numpy as np
filename = '/content/drive/MyDrive/Coding Dojo Data Science/PandasForDataManipulation/sales_predictions.csv'
df = pd.read_csv(filename)
df.head(20)
```

```python
# Renaming Columns
df.columns = ['Product ID', 'Weight','Low Fat or Regular','Visibility','Product Category','Product List Price','Store ID','Year Opened','Size of Store','Area of Store','Grocery Store or Supermarket','Item_Outlet_Sales']
df.head(20)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# Creating column "Size of Store 2" that displays "Size of Store" text respones as numbers
df.loc[df['Size of Store'] == 'High', 'Size of Store 2'] = 1
df.loc[df['Size of Store'] == 'Medium', 'Size of Store 2'] = 2
df.loc[df['Size of Store'] == 'Small', 'Size of Store 2'] = 3
df
```

```python
# Dropping "Size of Store" with text responses for now
df1 = df.drop(['Size of Store'], axis=1)
df1.head()
```

```python
df1.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'most_frequent')
df1 = imputer.fit_transform(df1)
```

```python
df1
```

```python
# Converting array above to dataframe
df1 = pd.DataFrame(df1, columns = ['Product ID', 'Weight','Low Fat or Regular','Visibility','Product Category','Product List Price','Store ID','Year Opened','Area of Store','Grocery Store or Supermarket','Item_Outlet_Sales','Size of Store 2'])
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
# Adding back column "Size of Store" that displays "Size of Store 2" impute calcualtion back as text
df1.loc[df1['Size of Store 2'] == 1, 'Size of Store'] = "High"
df1.loc[df1['Size of Store 2'] == 2, 'Size of Store'] = "Medium"
df1.loc[df1['Size of Store 2'] == 3, 'Size of Store'] = "Low"
df1
```
